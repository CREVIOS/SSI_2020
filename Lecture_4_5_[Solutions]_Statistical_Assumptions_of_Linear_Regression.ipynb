Create a scatterplot between height and weight from the dataset, and plot the line of best fit on the data. Also, calculate the predicted values. 

```
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
df = pd.read_csv('https://drive.google.com/u/0/uc?id=1UCtQoYaiN3DBgKUOn-BdrkYNXm_mCJT9&export=download')
plt.scatter(df["Height"], df["Weight"])
slope, intercept, r_value, p_value, std_err = stats.linregress(df["Height"], df["Weight"])
df["Weight_Pred"] = df["Height"] * slope + intercept
plt.plot(df["Height"], df["Weight_Pred"], c="orange")
```

Now plot the residuals, along with the x-axis. 

```
import numpy as np
plt.scatter(df["Height"], df["Weight"] - df["Weight_Pred"], c="red", marker=".")
plt.plot(df["Height"], np.zeros_like(df["Height"]), c="k")
```

Draw the QQ-plot of the residuals, and check that it's linear. 

```
import statsmodels.api as sm
sm.qqplot(df["Weight"] - df["Weight_Pred"])
```

Create a scatterplot between log Infant Mortality and log GDP from the dataset (after removing NA values), and plot the line of best fit on the data. Also, calculate the predicted values for log GDP given log Infant Mortality. 

```
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
df = pd.read_csv('https://drive.google.com/u/0/uc?id=1qizk0mqNHz6OCJoEvdJdQBLRcnrWUpOX&export=download')
df = df.dropna()

df["log_Infant_Mortality"] = np.log(df["Infant_Mortality"])
df["log_GDP"] = np.log(df["GDP"])
slope, intercept, r_value, p_value, std_err = stats.linregress(df["log_Infant_Mortality"], df["log_GDP"])
df["log_GDP_Pred"] = df["log_Infant_Mortality"] * slope + intercept

plt.scatter(df["log_Infant_Mortality"], df["log_GDP"], marker=".")
plt.plot(df["log_Infant_Mortality"], df["log_GDP_Pred"], c="orange")
plt.xlabel("log Infant Mortality")
plt.ylabel("log GDP")
```

Now plot the residuals, along with the x-axis. 

```
import numpy as np
plt.scatter(df["log_Infant_Mortality"], df["log_GDP"] - df["log_GDP_Pred"], c="red", marker=".")
plt.plot(df["log_Infant_Mortality"], np.zeros_like(df["log_Infant_Mortality"]), c="k")
```

Draw the QQ-plot of the residuals, and check that it's linear. 

```
import statsmodels.api as sm
sm.qqplot(df["log_GDP"] - df["log_GDP_Pred"])
```
